## Phase 2: Exploratory Data Analysis (EDA) & Data Cleaning

### 2.1. Data Cleaning (Handling Missing Values)

**a) Column count check for Missing Values**

We implemented the `count_missing` function to handle both NULLs and empty strings.

```python
def count_missing(c, dtype):
    if isinstance(dtype, StringType):
        return F.count(
            F.when(
                F.col(c).isNull() | (F.trim(F.col(c)) == ""), 
                c
            )
        ).alias(c)
    else:
        return F.count(F.when(F.col(c).isNull(), c)).alias(c)

missing_counts = df.select([
    count_missing(c, dtype) for c, dtype in df.dtypes
])

missing_data = missing_counts.collect()[0].asDict()

print(f"{'Column':<20} {'Missing Count':>15}")
print("-" * 40)
for col_name, count in missing_data.items():
    print(f"{col_name:<20} {count:>15}")
```

**b) Data Cleaning**

Before analyzing the data, we first handle invalid `TotalCharges` values by converting empty strings to null, then remove rows with missing `TotalCharges`, since they represent only a small number of records. Finally, we check for duplicate `customerID` entries to ensure data consistency before further preprocessing and modeling.

```python
# Handle TotalCharges: replace empty strings with null, then cast to Double
df = df.withColumn('TotalCharges', 
                   F.when(F.col('TotalCharges').isin("", " "), None)
                   .otherwise(F.col('TotalCharges')))

# Drop rows with null TotalCharges (11 records)
print(f"Rows before dropping null TotalCharges: {df.count()}")
df_clean = df.na.drop(subset=['TotalCharges'])
print(f"Rows after dropping null TotalCharges: {df_clean.count()}")

# Check for duplicates
print(f"Duplicate customerID count: {df_clean.count() - df_clean.select('customerID').distinct().count()}")
```

### 2.2. Data preparation

In this step, we ensure that all numerical features have the correct data types for modeling. Then we verify the updated schema to confirm the changes.

```python
df_clean = df_clean.withColumn('TotalCharges', F.col('TotalCharges').cast(DoubleType()))

# Cast SeniorCitizen to Integer (here the values are currently "0", "1")
df_clean = df_clean.withColumn('SeniorCitizen', F.col('SeniorCitizen').cast(IntegerType()))

# Cast tenure to Integer
df_clean = df_clean.withColumn('tenure', F.col('tenure').cast(IntegerType()))

# Cast MonthlyCharges to Double
df_clean = df_clean.withColumn('MonthlyCharges', F.col('MonthlyCharges').cast(DoubleType()))

print("Schema after conversion of columns:")
df_clean.printSchema()
```

### 2.3. Univariate Analysis (Analyzing Single Variables)

We compute statistics to understand the distribution of both numerical and categorical features.  
- For numerical variables, summary statistics such as mean, min, and max provide insight into scale and variability.
- For categorical variables (including SeniorCitizen and other key service-related features), we examine value counts to observe class balance and dominant categories.
  
This step helps us identify potential data imbalance, unusual distributions, and important patterns before modeling.

**a) Numerical Features**

```python
# numerical columns summary
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
df_clean.select(numerical_cols).describe().show()

# SeniorCitizen belongs to categorical column
print("SeniorCitizen Distribution:")
df_clean.groupBy('SeniorCitizen').count().show()
```

**b) Categorical Features**

```python
categorical_cols = [c for c in df_clean.columns if c not in numerical_cols + ['customerID', 'SeniorCitizen']]
print(f"Categorical columns: {categorical_cols}")

# Show value counts for key categorical columns
key_categorical = ['gender', 'Partner', 'Dependents', 'Contract', 
                   'InternetService', 'PaymentMethod', 'Churn']

for col in key_categorical:
    print(f"{col} distribution:")
    df_clean.groupBy(col).count().orderBy(F.desc('count')).show()
```

### 2.4. Bivariate Analysis (Analyzing Relationships)

We want to discover churn drivers through feature–target relationships. To do this, we compute churn rates across key categorical features such as Contract type and Internet Service, and compare numerical statistics between churned and non-churned customers.

```python
print("Churn rate by Contract type:")
df_clean.groupBy('Contract').agg(
    F.count('*').alias('total'),
    F.sum(F.when(F.col('Churn') == 'Yes', 1).otherwise(0)).alias('churned'),
    (F.sum(F.when(F.col('Churn') == 'Yes', 1).otherwise(0)) / F.count('*') * 100).alias('churn_rate')
).show()

print("Churn rate by Internet Service:")
df_clean.groupBy('InternetService').agg(
    F.count('*').alias('total'),
    F.sum(F.when(F.col('Churn') == 'Yes', 1).otherwise(0)).alias('churned'),
    (F.sum(F.when(F.col('Churn') == 'Yes', 1).otherwise(0)) / F.count('*') * 100).alias('churn_rate')
).show()

# Tenure statistics by Churn
print("Tenure statistics by Churn:")
df_clean.groupBy('Churn').agg(
    F.avg('tenure').alias('avg_tenure'),
    F.min('tenure').alias('min_tenure'),
    F.max('tenure').alias('max_tenure'),
    F.avg('MonthlyCharges').alias('avg_monthly_charges'),
    F.avg('TotalCharges').alias('avg_total_charges')
).show()
```

## Phase 3: Data Transformation & Feature Engineering

### 3.1. Identify Feature Columns

In this step, we exclude the identifier (`customerID`) and the target variable (`Churn`) from the feature set to prevent data leakage.  
The remaining columns are separated into categorical and numerical features to apply appropriate preprocessing techniques for each type.  
Finally, the target variable is converted into a binary numeric label (1 = churn, 0 = no churn) to make it compatible with machine learning algorithms.

```python
# we identify feature columns (except for customerID and target)
feature_cols = [c for c in df_clean.columns if c not in ['customerID', 'Churn']]

# we separate columns into categorical and numerical
cat_cols = [c for c in feature_cols if c not in numerical_cols]
num_cols = [c for c in feature_cols if c in numerical_cols]

print(f"Features to use: {feature_cols}")
print(f"Categorical: {cat_cols}")
print(f"Numerical: {num_cols}")

# we convert the target variable to numeric
df_clean = df_clean.withColumn('label', F.when(F.col('Churn') == 'Yes', 1).otherwise(0))
```

### 3.2. Define Pipeline Stages

**a) Categorical Encoding**

We encoded our categorical features in two steps: first converting strings to numbers with `StringIndexer`, then one-hot encoding with `dropLast=False` to keep all categories.

```python
indexers = []
encoders = []
indexed_cols = []
encoded_cols = []

for col in cat_cols:
    indexer = StringIndexer(inputCol=col, outputCol=f"{col}_indexed", handleInvalid='keep')
    encoder = OneHotEncoder(inputCol=f"{col}_indexed", outputCol=f"{col}_encoded", dropLast=False)
    indexers.append(indexer)
    encoders.append(encoder)
    indexed_cols.append(f"{col}_indexed")
    encoded_cols.append(f"{col}_encoded")
```

**b) Vector Assembly**

Combine numerical with encoded categorical

```python
# combine numerical features with encoded categorical features
assembler_inputs = num_cols + encoded_cols
vector_assembler = VectorAssembler(inputCols=assembler_inputs, outputCol='features_unscaled')
```

**c) Feature Scaling**

We apply **Feature Scaling** because our features combine one-hot encoded binaries (0/1) with unbounded numericals like `TotalCharges` (up to more than $8,000). Therefore, without scaling, the model would overweight dollar amounts and ignore service indicators.

```python
scaler = StandardScaler(inputCol='features_unscaled', outputCol='features', 
                        withStd=True, withMean=True)
```

**d) Create Pipeline**

We chained all preprocessing steps into a single **Pipeline**: 16 indexers, 16 encoders, 1 vector assembler, and 1 scaler. This ensures our transformations are fitted only on training data and applied consistently to test data, preventing data leakage and making the workflow reproducible.

```python
pipeline_stages = indexers + encoders + [vector_assembler, scaler]
pipeline = Pipeline(stages=pipeline_stages)

print(f"Pipeline stages: {len(indexers)} indexers, {len(encoders)} encoders, 1 assembler, 1 scaler")
print(f"Total stages in pipeline: {len(pipeline_stages)}")
```

## Phase 4: Linear Regression Classification

### 4.1. Define Logistic Regression & Imports

We configured **Logistic Regression** with standardized features and binary churn labels as our baseline classifier.

```python
lr = LogisticRegression(
    featuresCol='features',
    labelCol='label'
)
```

### 4.2. Build Full Pipeline (Preprocessing + LR)

We need to build complete pipeline having both stages Preprocessing and Modeling.

```python
lr_pipeline = Pipeline(stages=indexers + encoders + [vector_assembler, scaler, lr])

print('LR pipeline created with the following stages:')
for i, stage in enumerate(lr_pipeline.getStages()):
    print(f'  {i+1}. {stage.__class__.__name__}')
```

### 4.3. Train / Test Split

The dataset is split into 80% training and 20% test sets using a fixed random seed for reproducibility. The goal is to preserve the original class distribution across both sets to avoid bias during training and evaluation. After splitting, we verify the number of records and inspect the class distribution in the training set to ensure balance is maintained.

```python
train_df, test_df = df_clean.randomSplit([0.8, 0.2], seed=42)
print(f'Training set : {train_df.count()} records')
print(f'Test set     : {test_df.count()} records')
print('Class distribution in training set:')
train_df.groupBy('label').count().orderBy('label').show()
```

### 4.4. Define Hyperparameter Grid with ParamGridBuilder

We tune three key hyperparameters of Logistic Regression: the regularization strength (regParam) to control overfitting, the L1/L2 mixing ratio (elasticNetParam) to balance Ridge and Lasso penalties, and the maximum number of optimization iterations (maxIter) to ensure proper convergence. The grid explores multiple combinations to identify the best-performing model configuration.


```python
lr_param_grid = (
    ParamGridBuilder()
    .addGrid(lr.regParam,        [0.01, 0.1, 0.5])
    .addGrid(lr.elasticNetParam, [0.0, 0.5])
    .addGrid(lr.maxIter,         [10, 50])
    .build()
)

print(f'LR ParamGrid built — {len(lr_param_grid)} combinations')
for i, params in enumerate(lr_param_grid):
    combo = {p.name: v for p, v in params.items()}
    print(f'  Combo {i+1:>2}: {combo}')
```

### 4.5. Cross-Validation Setup (5-fold)

We use a `BinaryClassificationEvaluator` with **Area Under ROC (AUC)** as the evaluation metric. A 5-fold cross-validation is applied to ensure robust performance estimation. The `CrossValidator` tests all hyperparameter combinations defined in the grid and selects the configuration that achieves the highest average AUC.

```python
cv_evaluator = BinaryClassificationEvaluator(
    labelCol='label',
    rawPredictionCol='rawPrediction',
    metricName='areaUnderROC'
)

lr_cross_val = CrossValidator(
    estimator=lr_pipeline,
    estimatorParamMaps=lr_param_grid,
    evaluator=cv_evaluator,
    numFolds=5,
    seed=42
)
```

### 4.6. Train with Cross-Validation

We train the model using 5-fold cross-validation on all hyperparameter combinations. For each combination, we compute the average AUC-ROC across folds to evaluate classification performance. The results are printed in a structured table, and the configuration achieving the highest average AUC is then selected as the best model.

```python
lr_cv_model = lr_cross_val.fit(train_df)

print('\nAverage AUC-ROC per hyperparameter combination:')
print(f"{'Combo':<8} {'regParam':<10} {'elasticNet':<12} {'maxIter':<9} {'AUC-ROC':>8}")
print('-' * 53)
for i, (params, score) in enumerate(zip(lr_param_grid, lr_cv_model.avgMetrics)):
    combo = {p.name: v for p, v in params.items()}
    print(f"{i+1:<8} {combo['regParam']:<10} {combo['elasticNetParam']:<12} {combo['maxIter']:<9} {score:>8.4f}")

best_lr_score  = max(lr_cv_model.avgMetrics)
best_lr_idx    = lr_cv_model.avgMetrics.index(best_lr_score)
best_lr_params = {p.name: v for p, v in lr_param_grid[best_lr_idx].items()}
print(f'Best combo #{best_lr_idx+1}: {best_lr_params}  with  AUC = {best_lr_score:.4f}')
```

### 4.7. Evaluate Best Model on Test Set

The best cross-validated Logistic Regression model is applied to the test set to assess its generalization performance. We compute multiple evaluation metrics, including Accuracy, F1-score, Weighted Precision, Weighted Recall, and AUC-ROC, to provide a comprehensive performance assessment. Additionally, a confusion matrix is displayed to analyze the distribution of true vs. predicted classes and identify potential misclassification patterns.

```python
lr_predictions = lr_cv_model.transform(test_df)

lr_acc = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='accuracy').evaluate(lr_predictions)
lr_f1 = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='f1').evaluate(lr_predictions)
lr_prec = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='weightedPrecision').evaluate(lr_predictions)
lr_rec = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='weightedRecall').evaluate(lr_predictions)
lr_auc = BinaryClassificationEvaluator(labelCol='label', rawPredictionCol='rawPrediction', metricName='areaUnderROC').evaluate(lr_predictions)

print("Logistic Regression Evaluation")
print("-" * 40)
print(f"  Accuracy          : {lr_acc:.4f}")
print(f"  F1 Score          : {lr_f1:.4f}")
print(f"  Weighted Precision: {lr_prec:.4f}")
print(f"  Weighted Recall   : {lr_rec:.4f}")
print(f"  AUC-ROC           : {lr_auc:.4f}")

print("Confusion Matrix:")
lr_predictions.groupBy('label', 'prediction').count().orderBy('label', 'prediction').show()
```

## Phase 5: Random Forest Classification

### 5.1. Add RandomForestClassifier to the Pipeline

We added Random Forest to our pipeline, creating a 35-stage workflow that handles encoding, scaling, and classification in one reproducible flow.

```python
rf = RandomForestClassifier(
    featuresCol='features',
    labelCol='label',
    seed=42
)

rf_pipeline = Pipeline(stages=indexers + encoders + [vector_assembler, scaler, rf])

print("RF pipeline created with the following stages:")
for i, stage in enumerate(rf_pipeline.getStages()):
    print(f"  {i+1}. {stage.__class__.__name__}")
```

### 5.2. Define Hyperparameter Grid with ParamGridBuilder

We define a grid of hyperparameters to tune the **Random Forest** model. The grid searches across different numbers of trees, tree depths, and minimum samples per leaf node to balance model complexity and performance. In total, multiple hyperparameter combinations are evaluated to identify the best configuration.

```python
rf_param_grid = (
    ParamGridBuilder()
    .addGrid(rf.numTrees,             [50, 70])
    .addGrid(rf.maxDepth,             [4, 6, 8])
    .addGrid(rf.minInstancesPerNode,  [1, 2])
    .build()
)

print(f"ParamGrid built — {len(rf_param_grid)} hyperparameter combinations to evaluate")
for i, params in enumerate(rf_param_grid):
    combo = {p.name: v for p, v in params.items()}
    print(f"  Combo {i+1:>2}: {combo}")
```

### 5.3. Cross-Validation Setup (5-fold)

We apply 5-fold cross-validation to evaluate different Random Forest hyperparameter combinations. The process uses AUC-ROC as the evaluation metric to select the best model configuration. Cross-validation helps improve model robustness and reduce overfitting by testing performance across multiple data folds.

```python
rf_cross_val = CrossValidator(
    estimator=rf_pipeline,
    estimatorParamMaps=rf_param_grid,
    evaluator=cv_evaluator,
    numFolds=5,
    seed=42,
    parallelism=1
)
```

### 5.4. Train with Cross-Validation

The Random Forest model is trained using 5-fold cross-validation to evaluate different hyperparameter combinations. For each configuration, the average AUC-ROC score across validation folds is computed. The best-performing hyperparameter set is selected based on the highest average AUC-ROC score and used for final model evaluation.


```python
rf_cv_model = rf_cross_val.fit(train_df)

print("Average AUC-ROC per hyperparameter combination:")
print(f"{'Combo':<8} {'numTrees':<10} {'maxDepth':<10} {'minInst':<8} {'AUC-ROC':>8}")
print("-" * 50)
for i, (params, score) in enumerate(zip(rf_param_grid, rf_cv_model.avgMetrics)):
    combo = {p.name: v for p, v in params.items()}
    print(f"{i+1:<8} {combo['numTrees']:<10} {combo['maxDepth']:<10} {combo['minInstancesPerNode']:<8} {score:>8.4f}")

best_rf_score = max(rf_cv_model.avgMetrics)
best_rf_idx   = rf_cv_model.avgMetrics.index(best_rf_score)
best_rf_params = {p.name: v for p, v in rf_param_grid[best_rf_idx].items()}
print(f"Best combo #{best_rf_idx+1}: {best_rf_params}  with  AUC = {best_rf_score:.4f}")
```

### 5.5. Evaluate Best Model on Test Set

The best Random Forest model obtained from cross-validation is applied to the test set to evaluate its generalization performance. Multiple metrics are used, including Accuracy, F1-score, Weighted Precision, Weighted Recall, and AUC-ROC, to provide a comprehensive view of classification performance. A confusion matrix is also displayed to analyze correct and incorrect predictions across classes.

```python
rf_predictions = rf_cv_model.transform(test_df)

rf_acc = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='accuracy').evaluate(rf_predictions)
rf_f1 = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='f1').evaluate(rf_predictions)
rf_prec = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='weightedPrecision').evaluate(rf_predictions)
rf_rec = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='weightedRecall').evaluate(rf_predictions)
rf_auc = BinaryClassificationEvaluator(labelCol='label', rawPredictionCol='rawPrediction', metricName='areaUnderROC').evaluate(rf_predictions)

print("Random Forest Evaluation")
print("-" * 40)
print(f"  Accuracy          : {rf_acc:.4f}")
print(f"  F1 Score          : {rf_f1:.4f}")
print(f"  Weighted Precision: {rf_prec:.4f}")
print(f"  Weighted Recall   : {rf_rec:.4f}")
print(f"  AUC-ROC           : {rf_auc:.4f}")

print("Confusion Matrix:")
rf_predictions.groupBy('label', 'prediction').count().orderBy('label', 'prediction').show()
```

## Phase 5C: Advanced Model — GBTClassifier with Hyperparameter Tuning

*As a next step beyond **Logistic Regression** and **SVM**, we train a **Gradient Boosted Trees (GBT)** classifier.*  
*GBT builds trees sequentially — each tree corrects the errors of the previous one — making it much more powerful on non-linear tabular data like churn.*  

*We also apply **CrossValidator** to automatically find the best hyperparameter configuration via 5-fold cross-validation.*

### 5C.1. Define the GBT Model

Key GBT hyperparameters we will tune:
- **`maxIter`** — number of boosting rounds (trees)
- **`maxDepth`** — depth of each individual tree
- **`stepSize`** — learning rate (how much each tree contributes)

> **Note:** GBT in SparkML does not use `StandardScaler` features because tree-based models are scale-invariant.  
> So we have to rebuild the preprocessing pipeline **without** the scaler for GBT.

```python
# GBT does not need scaled features — rebuild assembler output directly
from pyspark.ml.feature import VectorAssembler

# New assembler: outputs 'features_gbt' (unscaled)
assembler_gbt = VectorAssembler(
    inputCols=num_cols + encoded_cols,
    outputCol='features_gbt'
)

gbt = GBTClassifier(
    featuresCol='features_gbt',
    labelCol='label',
    maxIter=50,        # default, will be tuned
    maxDepth=5,        # default, will be tuned
    stepSize=0.1,      # learning rate, will be tuned
    seed=42
)

print("GBTClassifier defined.")
```

### 5C.2. Assemble the GBT Pipeline

Again, we ssemble the pipeline StringIndexers -> OneHotEncoders -> VectorAssembler (unscaled) -> GBTClassifier.

```python
pipeline_gbt = Pipeline(
    stages=indexers + encoders + [assembler_gbt, gbt]
)

print("GBT Pipeline stages:")
for i, stage in enumerate(pipeline_gbt.getStages()):
    print(f"  {i+1}. {stage.__class__.__name__}")
```

### 5C.3. Define the Parameter Grid & CrossValidator

We search over a grid of `maxIter`, `maxDepth`, and `stepSize` combinations.  
CrossValidator runs **5-fold cross-validation** on each combination and selects the best.

```python
# Build parameter grid
paramGrid = (
    ParamGridBuilder()
    .addGrid(gbt.maxIter,   [20, 50])
    .addGrid(gbt.maxDepth,  [3, 5])
    .addGrid(gbt.stepSize,  [0.05, 0.1])
    .build()
)

print(f"Total parameter combinations to evaluate: {len(paramGrid)}")

# AUC-ROC evaluator — best metric for imbalanced binary classification
cv_evaluator = BinaryClassificationEvaluator(
    labelCol='label',
    rawPredictionCol='rawPrediction',
    metricName='areaUnderROC'
)

# CrossValidator
cv = CrossValidator(
    estimator=pipeline_gbt,
    estimatorParamMaps=paramGrid,
    evaluator=cv_evaluator,
    numFolds=5,
    seed=42,
    parallelism=2      # run 2 folds in parallel to speed things up
)

print("CrossValidator configured with 5-fold CV.")
```

### 5C.4. Train the CrossValidator

This fits the entire pipeline for every parameter combination across every fold.  
The best model is accessible via `cv_model.bestModel`.

```python
import time

print("Starting CrossValidator training — this may take a few minutes...")
t0 = time.time()

cv_model = cv.fit(train_df)

elapsed = time.time() - t0
print(f"Training complete in {elapsed:.1f}s ({elapsed/60:.1f} min)")
```

### 5C.5. Inspect the Best Hyperparameters

We extract the best configuration from the `CrossValidator`.

```python
best_pipeline_model = cv_model.bestModel

# The GBT model is the last stage of the best pipeline
best_gbt = best_pipeline_model.stages[-1]

print("Best hyperparameters found by CrossValidator:")
print(f"  maxIter  : {best_gbt.getNumTrees}")
print(f"  maxDepth : {best_gbt._java_obj.getMaxDepth()}")
print(f"  stepSize : {best_gbt._java_obj.getStepSize()}")

# Show CV scores for all combinations
print("\nAUC-ROC per parameter combination (mean across 5 folds):")
for score, params in sorted(zip(cv_model.avgMetrics, paramGrid), reverse=True):
    print(f"  AUC={score:.4f}  |  {params}")
```

### 5C.6. Evaluate the Best GBT Model on the Test Set

We run the best model on the held-out test set and compare directly against the LR baseline.

```python
# Predict on test set using best model
gbt_predictions = cv_model.transform(test_df)

print("Sample GBT predictions (top 10):")
gbt_predictions.select('customerID', 'Churn', 'label', 'prediction', 'probability') \
               .show(10, truncate=False)

# ── Metrics ──
gbt_acc = MulticlassClassificationEvaluator(
    labelCol='label', predictionCol='prediction', metricName='accuracy'
).evaluate(gbt_predictions)

gbt_f1 = MulticlassClassificationEvaluator(
    labelCol='label', predictionCol='prediction', metricName='f1'
).evaluate(gbt_predictions)

gbt_precision = MulticlassClassificationEvaluator(
    labelCol='label', predictionCol='prediction', metricName='weightedPrecision'
).evaluate(gbt_predictions)

gbt_recall = MulticlassClassificationEvaluator(
    labelCol='label', predictionCol='prediction', metricName='weightedRecall'
).evaluate(gbt_predictions)

gbt_auc = BinaryClassificationEvaluator(
    labelCol='label', rawPredictionCol='rawPrediction', metricName='areaUnderROC'
).evaluate(gbt_predictions)

print("-" * 40)
print(f"GBT Accuracy  = {gbt_acc:.4f}")
print(f"GBT F1 Score  = {gbt_f1:.4f}")
print(f"GBT Precision = {gbt_precision:.4f}")
print(f"GBT Recall    = {gbt_recall:.4f}")
print(f"GBT AUC-ROC   = {gbt_auc:.4f}")
print("-" * 40)

print("\nGBT Confusion Matrix:")
gbt_predictions.groupBy('label', 'prediction').count().orderBy('label', 'prediction').show()
```

### 5C.7. Direct Comparison: LR Baseline vs. GBT (Tuned)

We compared our tuned Gradient Boosted Trees against the Logistic Regression baseline and found GBT slightly underperformed across all metrics, with a -0.45% drop in AUC-ROC. So in this casethe simpler linear model generalizes better on this dataset, possibly due to the strong linear relationships in churn patterns.

```python
print("=" * 52)
print(f"  {'Metric':<16} {'LR Baseline':>14} {'GBT Tuned':>14}")
print("=" * 52)
print(f"  {'Accuracy':<16} {lr_acc:>14.4f} {gbt_acc:>14.4f}")
print(f"  {'F1 Score':<16} {lr_f1:>14.4f} {gbt_f1:>14.4f}")
print(f"  {'Precision':<16} {lr_prec:>14.4f} {gbt_precision:>14.4f}")
print(f"  {'Recall':<16} {lr_rec:>14.4f} {gbt_recall:>14.4f}")
print(f"  {'AUC-ROC':<16} {lr_auc:>14.4f} {gbt_auc:>14.4f}")
print("=" * 52)

improvement = (gbt_auc - lr_auc) / lr_auc * 100
print(f"\n  AUC-ROC improvement: {improvement:+.2f}% over baseline")
```

### 5C.8. Feature Importances

A key advantage of GBT over Logistic Regression: we can inspect which features the model relied on most.

```python
# Reconstruct feature names in the order VectorAssembler used them
feature_names = num_cols + encoded_cols

importances = best_gbt.featureImportances.toArray()

# Pair names and importances, sort descending
feat_imp = sorted(zip(feature_names, importances), key=lambda x: x[1], reverse=True)
top_n = 15
names_top, vals_top = zip(*feat_imp[:top_n])

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.barh(names_top[::-1], vals_top[::-1], color='#4c72b0', edgecolor='white')
ax.set_title(f'GBT — Top {top_n} Feature Importances', fontsize=13, fontweight='bold')
ax.set_xlabel('Importance Score')
plt.tight_layout()
plt.show()

print(f"\nTop 5 features:")
for name, val in feat_imp[:5]:
    print(f"  {name:<30} {val:.4f}")
```

### 6.1. Churn Distribution (Bar Plot)

**How balanced is our target variable?**  
Our target variable is imbalanced: 73.4% of customers stayed (No churn) while only 26.6% left (Yes churn). This 3:1 ratio isn't extreme but still requires attention. e'll rely on F1-score and AUC-ROC for fair evaluation and consider class weights or SMOTE if performance on the minority class suffers.

```python
churn_dist = df_clean.groupBy('Churn').count().toPandas()
churn_dist['pct'] = churn_dist['count'] / churn_dist['count'].sum() * 100

fig, ax = plt.subplots(figsize=(5, 4))
bars = ax.bar(churn_dist['Churn'], churn_dist['count'],
              color=['#4c72b0', '#dd8452'], edgecolor='white', linewidth=1.2)
for bar, (_, row) in zip(bars, churn_dist.iterrows()):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
            f"{row['pct']:.1f}%", ha='center', va='bottom', fontsize=11, fontweight='bold')
ax.set_title('Churn Distribution', fontsize=14, fontweight='bold')
ax.set_xlabel('Churn'); ax.set_ylabel('Number of Customers')
plt.tight_layout(); plt.show()
print(churn_dist.to_string(index=False))
```

### 6.2. Tenure vs. Churn (Box Plot)

Churned customers tend to have much shorter tenures — let's confirm visually.

```python
tenure_pd = df_clean.select('tenure', 'Churn').toPandas()

fig, ax = plt.subplots(figsize=(6, 4))
sns.boxplot(data=tenure_pd, x='Churn', y='tenure',
            palette={'No': '#4c72b0', 'Yes': '#dd8452'}, ax=ax)
ax.set_title('Tenure Distribution by Churn Status', fontsize=14, fontweight='bold')
ax.set_xlabel('Churn'); ax.set_ylabel('Tenure (months)')
plt.tight_layout(); plt.show()
```

### 6.5. Churn Rate by Internet Service Type (Horizontal Bar)

Fiber optic customers churn at almost 42%, making it far above DSL and no-internet customers.

```python
inet_churn = df_clean.groupBy('InternetService').agg(
    F.count('*').alias('total'),
    F.sum(F.when(F.col('Churn') == 'Yes', 1).otherwise(0)).alias('churned')
).withColumn('churn_rate', F.col('churned') / F.col('total') * 100).toPandas()
inet_churn = inet_churn.sort_values('churn_rate', ascending=True)

fig, ax = plt.subplots(figsize=(6, 3.5))
bars = ax.barh(inet_churn['InternetService'], inet_churn['churn_rate'],
               color='#dd8452', edgecolor='white')
for bar, val in zip(bars, inet_churn['churn_rate']):
    ax.text(val + 0.5, bar.get_y() + bar.get_height()/2,
            f"{val:.1f}%", va='center', fontsize=10)
ax.set_title('Churn Rate by Internet Service Type', fontsize=13, fontweight='bold')
ax.set_xlabel('Churn Rate (%)'); ax.set_xlim(0, 55)
plt.tight_layout(); plt.show()
```

### 6.6. Payment Method vs. Churn Rate (Grouped Bar)

We can see there is a strong correlation between payment method and churn because electronic check users show a striking 45.3% churn rate comparing with only other methods at below 20%.

```python
pay_churn = df_clean.groupBy('PaymentMethod').agg(
    F.count('*').alias('total'),
    F.sum(F.when(F.col('Churn') == 'Yes', 1).otherwise(0)).alias('churned')
).withColumn('churn_rate', F.col('churned') / F.col('total') * 100).toPandas()
pay_churn = pay_churn.sort_values('churn_rate', ascending=False)

fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.bar(pay_churn['PaymentMethod'], pay_churn['churn_rate'],
              color='#4c72b0', edgecolor='white')
for bar, val in zip(bars, pay_churn['churn_rate']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.4,
            f"{val:.1f}%", ha='center', fontsize=9)
ax.set_title('Churn Rate by Payment Method', fontsize=13, fontweight='bold')
ax.set_ylabel('Churn Rate (%)')
plt.xticks(rotation=15, ha='right')
plt.tight_layout(); plt.show()
```
